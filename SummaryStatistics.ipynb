```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
data = pd.read_csv("datasets/weight-height.csv")
data.head(5)
```

```python
data.shape
```

```python
# Describe function gives you a a quick statistical summary of your data
data.describe()
```

```python
height = data["Height"]
height.shape
```

```python
# One of the best visualization to use to view bunch of summary is BOX-plot

plt.figure(figsize = (12,8))
height.plot(kind="box", title = "Height Histogram")

# Central line within the box is median of your data
# The size of the box in a box-plot gives us the IQR, at the upper end is the 75% and lower end is 25% 
# The top WHISKER is q3+1.5(IQR) 
# The bottom WHISKER is q1-1.5(IQR)
# All points that are located outside the whisker can be outliers
```

```python
# KDE(Kernal density estimation) gives you a best view for Probability distribution curve
height.plot(kind="kde", title = "Height KDE", figsize = (12,8))
```

```python
plt.figure(figsize = (12,8))
height.plot(kind="hist", title = "Height Histogram")
# As you can see here the height data is bell shape normal distribution 
```

```python
weight = data["Weight"]
weight.shape
```

```python
weight.plot(kind="kde", title = "Weight KDE", figsize = (12,8))

# There are two peaks here, such distribution called as a bimodal distribution
# It's indicated that there are two categories or groups that might be present in dataset
```

```python
# Scatter plot for height and weight
plt.figure(figsize=(12,8))
sns.scatterplot(x='Height', y = 'Weight', data=data)
# It seems to be a clear Linear Regression in this data
```

```python
plt.figure(figsize=(12,8))
sns.scatterplot(x='Height', y = 'Weight',hue='Gender', data=data)
```

```python
# Group our data based on Gender
gender_groupby = data.groupby('Gender', as_index=False)
gender_groupby.head()
```

```python
gender_groupby.describe()
# Females are 0, males are 1
```
