```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sb
```

```python
height_weight_data = pd.read_csv("datasets/500_Person_Gender_Height_Weight_Index.csv")
height_weight_data.head(5)
```

```python
# Drop a column name Index
height_weight_data.drop('Index', inplace = True, axis = 1)
height_weight_data.shape
```

```python
# KDE(Kernal density estimation) gives you a best view for Probability distribution curve
height_weight_data[["Height"]].plot(kind="kde", title = "Height KDE", figsize = (12,8))
```

```python
height_weight_data[["Weight"]].plot(kind="kde", title = "Weight KDE", figsize = (12,8))
```

```python
# skew function can measure the skewness of your data
height_weight_data["Height"].skew()
## A measure of asymmetry around the mean can be positive or negative
## A value of 0 indicates no skew
```

```python
height_weight_data["Weight"].skew()
```

```python
# introduce some outliers
_data = [pd.Series(['Male', 205,460], index = height_weight_data.columns ),
        pd.Series(['Female', 202,390], index = height_weight_data.columns ),
        pd.Series(['Female', 199,410], index = height_weight_data.columns ),
        pd.Series(['Male', 202,390], index = height_weight_data.columns ),
        pd.Series(['Female', 199,410], index = height_weight_data.columns ),
        pd.Series(['Male', 201,460], index = height_weight_data.columns )
        ]
```

```python
height_weight_updated =  height_weight_data.append(_data, ignore_index = True)
height_weight_updated.tail()
```

```python
height_weight_updated[["Height"]].plot(kind="hist", bins=100, title = "Height", figsize = (12,8))
```

```python
height_weight_updated[["Weight"]].plot(kind="hist", bins=100, title = "Weight", figsize = (12,8))
```

```python
height_weight_updated[["Height"]].plot(kind="kde", title = "Weight KDE", figsize = (12,8))
```

```python
height_weight_updated[["Weight"]].plot(kind="kde", title = "Weight KDE", figsize = (12,8))
```

```python
height_weight_updated["Height"].skew()
```

```python
height_weight_updated["Weight"].skew()
```

```python
# Kurtosis- A measure of the likelihood of extreme events.
height_weight_data["Height"].kurtosis()

# The value is less than 3 indicating that in height distribution extreme events are likely than in a normal distribution
```

```python
height_weight_data["Weight"].kurtosis()
```

```python
height_weight_updated["Height"].kurtosis()
```

```python
height_weight_updated["Weight"].kurtosis()
# The likelihood of extreme events in Weight distribution is very very high
```
