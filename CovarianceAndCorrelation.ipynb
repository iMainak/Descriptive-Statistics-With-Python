```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
_data = pd.read_csv("datasets/automobile_data_processed.csv")
_data.head(5)
```

```python
# Best visualization for Bivariate is the scatterplot.
_data.plot.scatter(x='displacement', y='mpg', figsize=(12,8))

# It shows more powerful a car, lower its mileage
```

```python
# Horsepower vs MPG based on the number of cylinders that a car has

_data.plot.scatter(x='horsepower', y='mpg', c='cylinders', s=60,title="Automobile Data", figsize=(12,8))

# Graph shows:
## more horsepower less mileage
## more powerful car has more cylinders
##
```

```python
# cov function shows Covariance for all pairs of variables
_data.cov()
```

```python
# Great visulaization for correlation matrix is the heatmap

plt.figure(figsize=(12,8))
sns.heatmap(_data.corr(), annot=True)

# The color of each cell indicates the strength as well as the directionality of the crrelation
# Negative correlation use darker values
# Positive correlation are represented using lighter values
```
